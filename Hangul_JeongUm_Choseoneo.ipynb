```python
import os

if not os.path.exists('nl/한글_han'):
    os.mkdir('nl/한글_han')

for dir in sorted(hfns):
    fns = sorted(glob.glob(dir + '/*.txt'))
    for fn in fns:
        lines = open(fn).read().split('\n')
        filename = os.path.basename(fn)
        fout_name = os.path.basename(os.path.dirname(fn)).replace(' ', '') + filename[3:]
        print(fout_name)
        fout = open('nl/한글_han/' + fout_name, 'w', encoding='utf-8')
        try:
            for i, line in enumerate(lines):
                if line == '' or line.strip() == '': continue
                line = replace_var(line)      
                fout.write(line +'\n')
        except:
            print(f'error {i} line in {fn} file')
        fout.close()
```

```python
import konlpy
from konlpy.tag import Komoran
moran = Komoran(userdic="dic_komoran.user")
tokens = []
fout = open('hangul_tokens.txt', 'w', encoding='utf-8')
fns = sorted(glob.glob('nl/한글_han/*.txt'))
for fidx, fn in enumerate(fns): 
    lines = open(fn).read().split('\n')
    #fout.write(f"{fn} starts............\n")
    try:
        #for LDA, 1 line is corresponding to 1 document. 
        doc_lines = []
        for i, line in enumerate(lines):
            if line == '' or line.strip() == '' or line.startswith('<page>'): continue
            nrst = moran.nouns(line)
            tokens.extend(nrst)
            a_line = ' '.join(nrst)
            a_line = a_line.replace('애 용운동', '애용 운동')
            a_line = a_line.replace('월 천', '월간') # because Hanja typing errors (천, 간) in the source data
            doc_lines.append(a_line)
            #fout.write(a_line +'\n')
        doc_line = ' '.join(doc_lines)
        fout.write(doc_line +'\n')
    except:
        print(f'error {i} line in {fn} file')
        doc_lines =[]
        for i, line in enumerate(lines):
            if line.strip() == '': continue
            doc_lines.append(line)
        doc_line = ' '.join(lines)
        fout.write(doc_line +'\n')
print(tokens[:100])
print('num of pages', fidx+1)
```

```python
from gensim import corpora
import gensim
#import string
import pprint

''' Document-Term Matrix '''
# Turn our tokenized docs into a id-term dictionary.
dictionary = corpora.Dictionary(doc_term)
# Converting list of documents (corpus) into Document Term Matrix using above dictionary
doc_term_matrix = [dictionary.doc2bow(doc) for doc in doc_term]
```

```python
dictionary.__dict__ #2024.04.30
```

```python
''' Running LDA Model '''
n_topics= 9 # number of topic clusters ; 4,5,6,7,8,9,10,12,15
n_iterations=100 # how many times to run iterate through all doc data to get better clusters
lda_model = gensim.models.ldamodel.LdaModel(doc_term_matrix, id2word=dictionary,
                                            num_topics = n_topics, iterations=n_iterations,
                                            alpha='auto', eta='auto', random_state=1000,
                                            per_word_topics=True)

print('*** results ***')
n_words=15
pprint.pprint(lda_model.print_topics(num_topics=n_topics, num_words=n_words))
pprint.pprint(lda_model.get_topics())

# Save model, corpus, and dictionary in gensim lda formats to visualise
model_dir = f'hangul_lda_{n_topics}'
print(model_dir)
if not os.path.exists(model_dir):  
    os.mkdir(model_dir)

dictionary.save(model_dir +'/dictionary.dict')
corpora.MmCorpus.serialize(model_dir +'/corpus.mm',doc_term_matrix)
lda_model.save(model_dir +'/model.lda')
```

```python
f = open('clean_inputs.txt', 'r', encoding='utf-8') #after preprocessing of short terms (1 syllable terms). 
docs_list = f.readlines() 
print('# of documents', len(docs_list))
doc_term = [doc.split() for doc in docs_list]
print(doc_term[:5])
print(len(doc_term))

from gensim import corpora
import gensim
#import string
import pprint

''' Document-Term Matrix '''
# Turn our tokenized docs into a id-term dictionary.
dictionary = corpora.Dictionary(doc_term)
# Converting list of documents (corpus) into Document Term Matrix using above dictionary
doc_term_matrix = [dictionary.doc2bow(doc) for doc in doc_term]
dictionary.__dict__ #2024.04.30
```

### topic modelling in choseon

```python
''' Running LDA Model '''
n_topics= 6 # number of topic clusters ; 4,5,6,7,8,9,10,12,15
n_iterations=1000 # how many times to run iterate through all doc data to get better clusters
lda_model = gensim.models.ldamodel.LdaModel(doc_term_matrix, id2word=dictionary,
                                            num_topics = n_topics, iterations=n_iterations,
                                            alpha='auto', eta='auto', random_state=1000,
                                            per_word_topics=True)

print('*** results ***')
n_words=15
pprint.pprint(lda_model.print_topics(num_topics=n_topics, num_words=n_words))
pprint.pprint(lda_model.get_topics())

# Save model, corpus, and dictionary in gensim lda formats to visualise
model_dir = f'choseon_lda_{n_topics}'
print(model_dir)
import os
if not os.path.exists(model_dir):  
    os.mkdir(model_dir)

dictionary.save(model_dir +'/dictionary.dict')
corpora.MmCorpus.serialize(model_dir +'/corpus.mm',doc_term_matrix)
lda_model.save(model_dir +'/model.lda')
```
